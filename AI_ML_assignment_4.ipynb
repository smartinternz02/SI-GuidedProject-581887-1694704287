## Checking Outliers

```python
sns.boxplot(df["TotalWorkingYears"])
```

INFERENCE:  INFERENCE: Here We can see there are some outliers, We have to follow the outliers removing techniques and remove them

1.Inter Quatile Range(IQR Method)

2.Z-Score Method

3.Percentile Method

we can remove the outliers by using the any one of those methods.
Outlier removal by replacement with median

```python
q1=df.TotalWorkingYears.quantile(0.25)
q3=df.TotalWorkingYears.quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit=q1+1.5*IQR
```

```python
upper_limit
```

```python
median_TotalWorkingYears=df.TotalWorkingYears.median()
median_TotalWorkingYears
```

Handling the Outliers

```python
df["TotalWorkingYears"]=np.where(df["TotalWorkingYears"]>upper_limit,median_TotalWorkingYears,df["TotalWorkingYears"])
```

```python
sns.boxplot(df["TotalWorkingYears"])
```
